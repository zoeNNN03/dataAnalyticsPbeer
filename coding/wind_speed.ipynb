```python
import pandas as pd
import numpy as np
import os
```

```python
# โหลดข้อมูลจากไฟล์ Excel
df_main = pd.read_excel('../data/ความเร็วลม_พี่เจต.xlsx', sheet_name='Sheet1')

# drop duplicates rows on 'สถาณี' and 'เดือน/ปี'
df_main.drop_duplicates(subset=['สถานี', 'เดือน/ปี'], inplace=True)

df_main['เดือน/ปี'] = pd.to_datetime(df_main['เดือน/ปี'])
# แยกข้อมูลในคอลัมน์ 'เดือน/ปี' เพื่อสร้างคอลัมน์ 'Year' และ 'Month'
df_main['Year'] = df_main['เดือน/ปี'].dt.year
df_main['Month'] = df_main['เดือน/ปี'].dt.month
df_main.drop(['เดือน/ปี'], axis=1, inplace=True)
df_main.replace('-', np.nan, inplace=True)
df_main
```

```python
def get_day(row):
    if row['Year'] == 2020 and row['Month'] == 2:
        return 29
    elif row['Month'] in day30:
        return 30
    elif row['Month'] in day31:   
        return 31
    else:
        return 28

day31 = [1, 3, 5, 7, 8, 10, 12]
day30 = [4, 6, 9, 11]
df_main['countDay'] = df_main.apply(get_day, axis=1)

df_main.head()
```

```python
df = df_main.copy()

# สร้างคอลัมน์ใหม่ที่เก็บข้อมูลทั้งหมดในแต่ละแถวยกเว้นคอลัมน์ 'สถานี', 'Month', 'countDay'
# df['value'] = df.iloc[:, 1:-2].apply(list, axis=1)
# df['value'] = df.iloc[:, 1:df['countDay'].astype(int)].apply(list, axis=1)
df['value'] = df.apply(
    lambda row: row.iloc[1:int(row['countDay'])+1].tolist(), axis=1)

grouped_df = df.groupby(['สถานี', 'Year'])[
    'value'].apply(list).reset_index()

grouped_df['value'] = grouped_df['value'].apply(lambda x: np.concatenate(x))
grouped_df.head()
```

```python
# หาผลรวมของปริมาณน้ำฝนในสัปดาห์แรกของปี
# สร้าง dict ที่เก็บค่าเริ่มต้นของสัปดาห์แรกของแต่ละปี
year_data = {
    2013: 5,
    2014: 4,
    2015: 3,
    2016: 2,
    2017: 7,
    2018: 6,
    2019: 5,
    2020: 4,
    2021: 2,
    2022: 1
}


def find_avg_of_start_year(row):
    year = row['Year']
    start_week = year_data[year]

    # ตรวจสอบและแทนที่ค่า '-' ด้วย np.nan
    numeric_values = [np.nan if value ==
                      '-' else float(value) for value in row['value'][:start_week]]
    
    # คำนวณค่าเฉลี่ยของตัวเลขไม่เอา NaN
    numeric_values = [value for value in numeric_values if not np.isnan(value)]
    avg_value = np.sum(numeric_values)
    return row['สถานี'], row['Year'], 1, avg_value

start_year = pd.DataFrame()
start_year['สถานี'], start_year['Year'], start_year['start_of_week'], start_year['wind_speed'] = zip(
    *grouped_df.apply(find_avg_of_start_year, axis=1))
start_year
```

```python
# ทำสัปดาห์อื่นๆ นอกจากสัปดาห์ที่ 1
# grouped_df['value'] = grouped_df['value'].apply(lambda x: x[7:])
# grouped_df.head()

def drop_data(row):
    year = row['Year']
    start_week = year_data[year]
    row['value'] = row['value'][start_week:]
    return row


grouped_df = grouped_df.apply(drop_data, axis=1)
grouped_df
```

```python
# นำข้อมูลทั้งหมดมาหาผลรวมเป็นรายสัปดาห์ แล้วนำมารวมกันในคอลัมน์ 'weekly_average'
new_df = grouped_df.copy()
def calculate_weekly_average(values):
    numeric_values = list(map(float, values))
    # แบ่งค่าเป็นกลุ่มของรายสัปดาห์
    weekly_values = [numeric_values[i:i+7] for i in range(0, len(numeric_values), 7)]
    # drop nan values
    weekly_values = [[val for val in sublist if not np.isnan(
        val)] for sublist in weekly_values]
    # หาผลรวมของแต่ละสัปดาห์
    weekly_averages = [np.sum(weekly) for weekly in weekly_values]
    return weekly_averages

new_df['weekly_average'] = new_df['value'].apply(calculate_weekly_average)
new_df.drop(['value'], axis=1, inplace=True)
new_df.head()
```

```python
# แยกข้อมูลออกจากกันในคอลัมน์ 'weekly_average' และสร้าง DataFrame ใหม่เพื่อนำมาเก็บข้อมูลใหม่ที่ทำการเพิ่มลำดับของสัปดาห์เริ่มต้น
new_data = []

# วนลูปเพื่อแบ่งข้อมูลในคอลัมน์ 'weekly_average' และสร้างข้อมูลใหม่
for index, row in new_df.iterrows():
    station = row['สถานี']
    year = row['Year']
    weekly_averages = row['weekly_average']

    # วนลูปเพื่อสร้างข้อมูลใหม่สำหรับแต่ละค่าในคอลัมน์ 'weekly_average'
    for i, avg in enumerate(weekly_averages):
        # คำนวณหาค่าของสัปดาห์เริ่มต้น
        if i == 0:
            new_data.append([station, year, 1, np.nan])
        start_of_week = i + 2
        new_data.append([station, year, start_of_week, avg])

# สร้าง DataFrame จากข้อมูลใหม่
new_df = pd.DataFrame(
    new_data, columns=['สถานี', 'Year', 'start_of_week', 'wind_speed'])

# แสดง DataFrame ที่ได้
# new_df.to_csv(f'./dataset/temp/temp2013-2022.csv', index=False, encoding='utf-8-sig')
new_df.head()
```

```python
result_df = new_df.merge(
    start_year, on=['สถานี', 'Year', 'start_of_week'], how='left', suffixes=('_df1', '_df2'))
result_df['wind_speed_df1'].fillna(result_df['wind_speed_df2'], inplace=True)
result_df.drop('wind_speed_df2', axis=1, inplace=True)
result_df.rename(columns={'wind_speed_df1': 'wind_speed'}, inplace=True)
result_df.head()
```
