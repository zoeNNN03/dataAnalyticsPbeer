# รวมข้อมูลเป็นรายปี และหาค่าเฉลี่ยเป็นรายวัน

```python
# รวมข้อมูลเป็นรายเดือน
years = ['2012', '2013', '2014', '2015', '2016', '2017', '2018', '2019', '2020', '2021', '2022', '2023']
months = ['01', '02', '03', '04', '05', '06', '07', '08', '09', '10', '11', '12']

for year in years:
    for month in months:
        folder_path = f'./downloaded_csv/{year}/{year}{month}/raw'
        # สร้าง DataFrame ที่จะรวมข้อมูล
        combined_df = pd.DataFrame()
        
        # วนลูปอ่านและรวมข้อมูลจากไฟล์ .csv ใน folder
        for filename in os.listdir(folder_path):
            if filename.endswith('.csv'):
                file_path = os.path.join(folder_path, filename)
                df = pd.read_csv(file_path)
                # combined_df = combined_df.append(df, ignore_index=True)
                combined_df = pd.concat([combined_df, df], ignore_index=True)
        # Export ไปเป็นไฟล์ .csv
        # พิมพ์ปี เดือนที่กำลังอ่าน
        print(f'{year} {month}')
        combined_df.to_csv(
            f'./downloaded_csv/{year}/{year}{month}/{year}{month}.csv', index=False)
```

```python
# รวมข้อมูลเป็นรายปี
years = ['2012', '2013', '2014', '2015', '2016', '2017',
         '2018', '2019', '2020', '2021', '2022', '2023']
months = ['01', '02', '03', '04', '05',
          '06', '07', '08', '09', '10', '11', '12']

for year in years:
    # สร้าง DataFrame ที่จะรวมข้อมูล
    combined_df = pd.DataFrame()
    for month in months:
        folder_path = f'./downloaded_csv/{year}/{year}{month}'
        # วนลูปอ่านและรวมข้อมูลจากไฟล์ .csv ใน folder
        for filename in os.listdir(folder_path):
            if filename.endswith('.csv'):
                # พิมพ์ไฟล์ที่กำลังอ่าน
                print(f'{filename}')
                file_path = os.path.join(folder_path, filename)
                df = pd.read_csv(file_path)
                # combined_df = combined_df.append(df, ignore_index=True)
                combined_df = pd.concat([combined_df, df], ignore_index=True)
    # Export ไปเป็นไฟล์ .csv
    combined_df.to_csv(
            f'./downloaded_csv/{year}/{year}.csv', index=False)
```

```python
years = ['2012', '2013', '2014', '2015', '2016', '2017',
         '2018', '2019', '2020', '2021', '2022', '2023']

for year in years:
    print(year)
    folder_path = f'./downloaded_csv/{year}/{year}.csv'
    df = pd.read_csv(folder_path)

    # แทนที่ค่า -999.0 ด้วย NaN
    df['temp'].replace([-999.0, "-"], pd.NA, inplace=True)

    # แปลงคอลัมน์ 'temp' เป็น numeric
    df['temp'] = pd.to_numeric(df['temp'], errors='coerce')

    # แปลงคอลัมน์ 'date' เป็น datetime
    df['date'] = pd.to_datetime(df['date'])

    # กรองข้อมูลที่มีค่า temp เป็น NaN
    df = df[df['temp'].notna()]

    # หาค่าเฉลี่ย
    average_temp = df.groupby(['date', 'code'])['temp'].mean().reset_index()
    # Export ไปเป็นไฟล์ .csv
    average_temp.to_csv(
        f'./downloaded_csv/{year}/{year}_average.csv', index=False)
```

# หาค่าเฉลี่ยเป็นรายสัปดาห์ และรวมคอลัมน์จาก index.csv

```python
years = ['2012', '2013', '2014', '2015', '2016', '2017',
         '2018', '2019', '2020', '2021', '2022', '2023']

for year in years:
    print(year)
    folder_path = f'./downloaded_csv/{year}/{year}_average.csv'
    df = pd.read_csv(folder_path)

    # แปลงคอลัมน์ 'date' เป็น datetime
    df['date'] = pd.to_datetime(df['date'])

    # เพิ่มคอลัมน์ 'start_of_week' ที่แสดงวันที่แรกของสัปดาห์
    # df['start_of_week'] = df.groupby(['code', 'date'])['date'].transform(
    #     lambda x: x - pd.to_timedelta(x.dt.dayofweek, unit='D')).dt.strftime('%Y-%m-%d')
    # df.insert(0, 'start_of_week', df.pop('start_of_week'))

    # เพิ่มคอลัมน์ 'week_of_year' ที่แสดงสัปดาห์ของปี
    df['week_of_year'] = df['date'].dt.strftime(
        '%U').astype(int) + 1
    
    # หาค่าเฉลี่ยรายสัปดาห์
    df = df.groupby(['week_of_year', 'code']).agg({
        'temp': 'mean'
    }).reset_index()
    
    # Export ไปเป็นไฟล์ .csv
    df.to_csv(
        f'./downloaded_csv/{year}/{year}_weekofyear.csv', index=False)
```

# Request2: รวมข้อมูลเป็นจังหวัด

```python
# loop file in dataset folder follow year
years = ['2012', '2013', '2014', '2015', '2016', '2017', '2018', '2019', '2020', '2021', '2022', '2023']
for year in years:
    file_name = './dataset/version1/' + year + '_final.csv'
    df = pd.read_csv(file_name)
    df.dropna(subset=['province'], inplace=True)
    df.drop(['code', 'name', 'lat', 'long', 'tambon', 'amphoe', 'basin', 'region',
            'regionEng', 'station_type'], inplace=True, axis=1)

    # หาค่าเฉลี่ยตามจังหวัด

    data = df.groupby(['week_of_year', 'province']).agg({'temp': 'mean'}).reset_index()
    data['provinceEng'] = data['province'].map(province)
    data['year'] = year
    data = data.reindex(['year', 'week_of_year', 'province', 'provinceEng', 'temp'], axis="columns")
    data.to_csv(
        f'./dataset/version2/{year}_province.csv', index=False, encoding='utf-8-sig')
```

```python
import pandas as pd
import os
directory_path = './downloaded_csv/'
# กำหนดคอลัมน์ที่ต้องการหาค่าเฉลี่ย
value_column = 'temp'

if os.path.exists(directory_path):
    for file in os.listdir(directory_path):
        if file.endswith('.csv'):
            # กำหนด path ของไฟล์ CSV
            file_path = f'2012.csv'
            df = pd.read_csv(file_path)

            # ลบแถวที่คอลัมน์ 'temp' มีค่าเป็น -999.0 หรือ '-'
            df = df[(df['temp'] != -999.0) & (df['temp'] != '-')]
            df['date'] = pd.to_datetime(df['date'])

            print(file)

            # ทำการ group by ตามคอลัมน์ 'date' และคอลัมน์ 'code' และคำนวณค่าเฉลี่ย
            grouped_main = df.groupby(['date', 'code'])[value_column].mean().reset_index(name='temp')

            # สร้าง DataFrame ที่มีทุกรายการของ 'date' และ 'code'
            date_range = pd.date_range(df['date'].min(), df['date'].max(), freq='D')
            all_codes = df['code'].unique()
            all_combinations = pd.MultiIndex.from_product([date_range, all_codes], names=['date', 'code'])
            result_df = pd.DataFrame(index=all_combinations).reset_index()

            # Merge DataFrame เพื่อเติมค่าเฉลี่ยลงในตาราง
            result_df = pd.merge(result_df, grouped_main, on=['date', 'code'], how='left')

            # เติมค่า 0 แทน NaN
            # result_df[value_column].fillna(0, inplace=True)

            # drop nan
            result_df.dropna(subset=['temp'], inplace=True)

            # เพิ่มคอลัมน์ 'start_of_week' ที่แสดงวันที่แรกของสัปดาห์
            result_df['start_of_week'] = result_df.groupby(['code', 'date'])['date'].transform(lambda x: x - pd.to_timedelta(x.dt.dayofweek, unit='D')).dt.strftime('%Y-%m-%d')

            # เพิ่มคอลัมน์ 'week_of_year' ที่แสดงสัปดาห์ของปี
            result_df['week_of_year'] = result_df['date'].dt.strftime('%U').astype(int) + 1

            # copy data
            df_2 = result_df.copy()

            # กำหนดคอลัมน์ที่ต้องการหาค่าเฉลี่ย
            value_column = 'temp'

            # ทำการ group by ตามคอลัมน์ 'date' และคำนวณค่าเฉลี่ย
            data_start_of_week = df_2.groupby(['start_of_week', 'code'])[value_column].mean()

            # รวม table และ index
            data_start_of_week = pd.merge(df_2, index_, on='code', how='left')

            # drop col 'stare_of_week'
            data_start_of_week.drop(columns=['start_of_week'], inplace=True)

            # result
            data_start_of_week.to_csv(f'dataset2/data_start_of_week{file}', index=False, encoding='utf-8-sig')

            print(data_start_of_week)

else:
    print(f"The directory {directory_path} does not exist.")
```
